```python
# Importing time to cause delay between functions.
import time

#Importing random to randomize coin toss.
import random

#Importing sys to force exit game when there is error.
import sys

# Defining fail function. If the the player chooses not to continue the game will end and exit after pressing enter.
def fail():
    
    print("""
GAME OVER!!! Thank you for playing!""")
    input(prompt="""
Press enter to exit""")
    exit()  # Exit the program

# Defining win function. After winning all the stages, the player will press enter to exit from the game.
def win():
    print("""
Congratulations! You have successfully completed the adventure. Thank you for playing!""")
    input(prompt="""
Press enter to exit""")
    sys.exit()  # Exit the program

# Introduction and instruction of the game.
def stage_0():

    # The gaming instruction.
    print("""
Welcome to the adventure of Harry Potter and the Philosopher's Stone. Before we start please read carefully 
the following instructions.
    
1. Throughout the game please follow the instructions to input and make choices.
2. To win the game, you have to pass all three stages with Ron and Hermione. If any of you die or get caught 
by anyone, the game will be over.
3. If the game is over, you may start from the latest checkpoint, which is the beginning of each stage.

References

Rowling, J. K. (1997). *Harry Potter and the Philosopher's Stone*. Bloomsbury Publishing.""")
    
    print("_" * 130) #updated the coding of long line, which is for separating stages.

    # Pause execution for 5 seconds to give the user time to read instructions.

    time.sleep(5)

    # To give a brief introduction story for the user.
    print("""
Let's rewind the previous story.""")
    
    time.sleep(2)

    print("""
It’s your first year at Hogwarts School of Witchcraft and Wizardry. Strange and unsettling occurrences have 
been happening on campus such as a troll attack, and your magical broom incident during a Quidditch match. 
Clearly, something dark and threatening is lurking around.""")

    time.sleep(3)

    print("""
After you visit the Forbidden Forest with Draco, Hagrid, and Fang, you find out something sinister is preying 
on the unicorn blood to survive. All the clues were leading to Professor Snape. Despite your warnings to 
Professor McGonagall that Professor Snape is after the fabled Philosopher's Stone, the professor didn’t take 
your concerns seriously.""")

    time.sleep(3)

    print("""
Now, you must take matters into your own hands. You are determined to find the Philosopher's Stone before the 
mysterious Professor Snape does, because if he succeeds, the Dark Lord would return and the wizarding world could 
be catastrophic.""")

    # Continue with the next part of the story
    print("""
You gather your courage and prepare to set out on your quest...""")
	
    # Function to continue from stage_1
    stage_1()

# Function for stage 1 of the game
def stage_1():
    # Introduction to Stage 1
    print("""
STAGE 1""")
    time.sleep(2)
    
    print("""
Ron: The door is locked. How are we supposed to open the door?
Hermione: Let's use a spell to open the door. Which spell shall we use, Harry?""")
    
    # Present the list of options of spells to the user
    # List of spells
    spells = [
        ("A", "Alohomora", "Unlocking spell"),
        ("B", "Bombarda", "Small detonation spell"),
        ("C", "Lumos", "Light spell"),
    ]

    # Shuffle the spells
    random.shuffle(spells)

    # Display randomized options
    print("Choose a spell:")
    for option in spells:
        print(f"{option[0]}. {option[1]}: {option[2]}")

    # Loop to handle user input and errors.
    while True:
        try:
            # To get the player's answer for choice (either letter or spell name)
            choice = input(prompt="""
Enter the letter of the spell or the spell name (Alohomora, Bombarda, Lumos)\n""").strip().lower()

            # Outcomes based on the chosen spell
            if choice in ["a", "alohomora", "alohomor", "aloha"]: 
                print("""
You cast Alohomora, and the door unlocks smoothly! You may proceed.
""")
                time.sleep(3)
                
                print("Door creaking sound...\n")
                time.sleep(1)

                #  Continuation of story as listed below. 
                story_sentences = [
                    "Hermione: Shhhh, the enchanted harp is still playing the music. Fluffy must be sleeping. Let's slowly head to the door.",
                    "Ron: Oh no, his paw is on the door.",
                    "Hermione: Let's move it.",
                    "All three of you slowly moved his paw and opened the door, but you noticed the harp was not playing lullaby anymore.",
                    "Harry: Guys, why is it so silent here?",
                    "Suddenly, big and slimy liquid drooled over Ron's shoulder. When the three of you looked up, Fluffy was standing there looking "
					"into your eyes and about to eat you alive. The three of you screamed and quickly jumped to the door."
                ]

                # Loop through the sentences one by one pressing enter until it finishes.
                for sentence in story_sentences:
                    print(sentence)
                    input(prompt="Press Enter to continue...\n")  # Wait for the user to press Enter before showing the next sentence.

                # Function to visualize the game scenario.
                print("Will you look up to see what Fluffy is doing?")
                print("A. Yes, we are safe anyway.")
                print("B. No, I am too scared to see that scenario.\n")

                # Loop to handle input
                while True:
                    # Get the player's answer for choice_fluffy (letter or yes/no)
                    choice_fluffy = input(prompt="Enter the letter (a/b) or (yes/no):\n").strip().lower()

                    if choice_fluffy in ["yes","a", "y"]:  
                        print("\nYou see Fluffy barking and throwing away the door with just one bite. Luckily, he was too big for the door.\n")
                        break  # Exit the loop if a valid choice is made
                        
                    elif choice_fluffy in ["no", "n", "b"]: 
                        print("\nYou closed your eyes tightly because you were too scared to look up. All you could hear was Fluffy barking loud and clear, but as you crouched down, the barking was getting lower.\n")
                        break  # Exit the loop if a valid choice is made
                        
                    else:print("Wrong input. Please enter the letter of the answer (A/B) or Yes/No.") # To handle other inputs from the user.

                # Ending stage_1 and proceeding to stage_2
                print("Congratulations! You have completed Stage 1.")
                stage_2()  # Continue to stage 2
                break  # Move to the next stage if needed
                
            elif choice in ["bombarda", "b", "bomb", "bombard"]:  
                time.sleep(2) 
                print("""
You cast Bombarda! The door is blown off its hinges with a loud bang.
Suddenly, Mr. Filch appears and catches you red-handed.
Game Over! You've been caught and sent to the dungeons to reflect on what you have done.""")
                
                # Giving option to the user to restart from stage 1.
                restart_choice_1 = input(prompt="Do you want to restart the game? (yes/no):\n").strip().lower()
                if restart_choice_1 in ["yes", "y"]:  
                    stage_1()  # Restart the game by calling the spell casting function again.
                    break
                    
                elif restart_choice_1 in ["no", "n"]: 
                    fail() # End the game by returning to fail function.
                    return
                
                else:
                    print("Invalid choice. Please enter 'yes' or 'no'.\n")
                    
            elif choice in ["lumos", "c", "lum"]: 
                print("You cast Lumos. A soft light appears, but it doesn’t help unlock the door. Try again!\n")
                
            else:print("That's not a valid choice. Please choose A, B, C, or type the spell name.\n")
        except Exception as e:
            print(f"An error occurred: {e}. Please try again.")
                
    # Stage 2 continuation
             
def stage_2():
    
    devil_snare_hp = 100  # Starting HP of Devil's Snare/enemy.
    chances = 2  # Player has two chances to defeat Devil's Snare.
    
    time.sleep(3)
    
    print("_" * 130) #updated the coding of long line, which is for separating stages.
    
    print("""
STAGE 2
""")
    time.sleep(2)

    # Stage_2 Introduction story
    story_intro = (
        "You slide through dark and smelly place as if it was some sort of pipe, "
        "you land softly. When you look around, you find yourself surrounded by black roots. "
        "Suddenly, the roots start wrapping around your bodies. " 
        "Hermione: 'Stop moving, both of you! This is Devil's Snare! You have to relax. "
        "If you don't, it'll only kill you faster!\n"
        "Hermione stops moving, and slides down to the ground with a 'See, I told you' "
        "expression, leaving Ron and you screaming."
    )

    # Print the story in one chunk
    print(story_intro)
    input(prompt="Press Enter to continue...\n") 
    
    # Ask player if they calm down
    print("""
Will you follow Hermione's advice and calm down?
A. Yes, of course. She sounds safe and sound, unless someone is mimicking her voice.
B. NO, WHAT DO YOU MEAN TO CALM DOWN? I have to get out!
""")

    while True:
        choice_devils_snare = input(prompt="Enter the letter (A/B) or Yes/No:\n").strip().lower()
        if choice_devils_snare in ["a", "yes", "y"] : 
            print("""
You stopped struggling and sank down where you meet Hermione. But Ron is still up there screaming. 
Hermione remembers that Devil's Snare hates light and warmth.
""")
            break # Ending the loop and continue to next function.
            
        elif choice_devils_snare in ["b","no", "n"]:  
            print("Each time you struggle, the roots are wrapping even tighter. You feel your vision dimming as you slowly close your eyes.\n")
            time.sleep(2)
            print("GAME OVER!!!")
            
            restart_choice_2 = input(prompt="Do you want to continue from the latest checkpoint? (yes/no):\n ").strip().lower()
            if restart_choice_2 in ["yes", "a", "y"]:  
                stage_2()  # Restart the game by calling the letter or yes/no function again
                return 
            
            elif restart_choice_2 in ["no", "b", "n"] :
                fail()
                return # Heading to Fail function to exit the game.
            
            else:print("Invalid choice. Please enter 'yes' or 'no'.") #To show error if other inputs are entered. 
                
        else:print("Invalid choice. Please enter A or B.")
    
    while chances > 0 and devil_snare_hp > 0:
        print(f"\nDevil's Snare HP: {devil_snare_hp}")
        print("Choose your action:")
        print("A. Lumos: Creates dim light.")
        print("B. Lumos Maxima: Creates much stronger light.")
        print("C. Hermione, HELP!!!")

        choice = input(prompt="Enter your choice (A, B, or C):\n").strip().lower()  
        

        if choice in ['a', 'lumos']:         
            devil_snare_hp -= 1
            chances -= 1
            print("\nYou cast Lumos. A dim light weakens the Devil's Snare slightly.")
        elif choice in ['b', 'lumos maxima']:
            devil_snare_hp -= 30
            chances -= 1
            print("You cast Lumos Maxima. A much stronger light weakens the Devil's Snare significantly.")
        elif choice in ['c', 'lumos solem']: 
            print("\nHermione casts Lumos Solem, creating sunlight! The Devil's Snare's HP drops to 0.")
            devil_snare_hp = 0
            break
        else:
            print("Invalid choice. Please enter A, B, or C.\n")
            continue

        if devil_snare_hp <= 0:
            break

        print(f"\nYou have {chances} chance(s) left.")
        if chances == 0:
            print("Ron is still trapped! The Devil's Snare squeezes tighter, and he dies.\n")
            print("GAME OVER!!!")
            restart_choice_3 = input(prompt="Do you want to continue from stage 2? (yes/no): ").strip().lower()
            if restart_choice_3 in ["yes", "y"]:
                stage_2()  # Restart the game
                return
            elif restart_choice_3 in ["no", "n"]: 
                fail()  # Exit the game
                return
            else:
                print("Invalid input. Exiting game.")
                return
        else: pass  # 

    print("Congratulations! You have defeated the Devil's Snare and saved Ron. You passed Stage 2!")
    stage_3()  # Continue to stage 3
          
    # stage_3()
    

def stage_3():
    print("_" * 130) #updated the coding of long line, which is for separating stages.

    print("""
FINALE STAGE""")
    
    # Countdown to start the stage.
    countdown = [3, 2, 1]
    for number in countdown:
        print(number)
        time.sleep(1)
    time.sleep(3)

    # Story for the final stage
    print("""
After saving Ron, the three of you continue your journey to protect the Philosopher's Stone. You have encountered
two pathways. One says "Tails", and the other one says "Heads". But there was no road to reach the other side.
    """)

    print("You see a coin on an altar...\n")
    time.sleep(3)

    # Let the player choose a method to reach the other side
    print("Will you toss the coin?")
    print("A. Yes, of course.")
    print("B. No, let's just use Wingardium Leviosa to levitate each other.\n")

    while True:
        choice_pass_cliff = input(prompt="Enter the letter (A/B) or Yes/No: ").strip().lower() 

        if choice_pass_cliff in ['a', 'yes', 'y']:   
            print("You took the coin and flipped it.\n")
            time.sleep(3) # Delaying next function to give time for coin flip.

            # Coin toss result randomizing.
            coin_result = random.choice(["Heads", "Tails"])
            time.sleep(3) # Delaying next function to give the result for coin flip.
            print("The coin toss result is:", coin_result)

            if coin_result == "Heads":
                print("""
It is Heads. Suddenly, there is an earthquake. A mass of soil forms together and creates a bridge to the Heads path. """)
                print("To be continue...\n")
                win()  #Ending the game by going to win.
                return
            
            elif coin_result == "Tails":
                print("""
It is Tails. Suddenly, one of the big tiles on the ground lifts up, revealing three magic brooms.
Each of you each grabs a broom and flies toward the Tails path.
                """)
                print("To be continue...\n")
                win() #Ending the game by going to win.
                return
            
            else: 
                print("Unexpected result. Retossing the coin...\n")
                time.sleep(2)
                
        elif choice_pass_cliff in ['b', 'no', 'n']:
            time.sleep(2)
            print("\nRon: Wingardium LeviosA.") 
            time.sleep(2)
            print("""
Ron tries to levitate you first, but he drops you due to incorrect pronunciation.
It's Wingardium LeviOsa, not Wingardium LeviosA. 
            """) #Not error- LeviosA 
            print("GAME OVER!!!\n")
            time.sleep(2)

            restart_choice_4 = input(prompt="Do you want to continue from the last saved checkpoint? (yes/no): ").strip().lower() 
            if restart_choice_4 in ["yes", "y"]:
                stage_3()  # Restart the game by yes.
                return
            
            elif restart_choice_4 in ["no", "n"]:
                fail() # If the player chooses no, the game will go to fail function and exit 
                return
            
            else:
                print("Invalid choice. Please enter 'yes' or 'no'.")
                
        else:
            print("Invalid choice. Please enter A or B.")
stage_0()
```
